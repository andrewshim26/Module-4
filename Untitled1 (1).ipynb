```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
```

```python
# Load the dataset
file_path = 'Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products.csv'  
data = pd.read_csv(file_path)

# Select relevant columns and drop missing values
reviews = data[['reviews.text', 'reviews.rating']].dropna()

# Preprocess reviews text
def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    text = ''.join([char for char in text if char.isalnum() or char.isspace()])  # Remove special characters
    return text

reviews['cleaned_text'] = reviews['reviews.text'].apply(preprocess_text)

# Convert text data to TF-IDF representation
tfidf = TfidfVectorizer(stop_words='english', max_features=500)
X = tfidf.fit_transform(reviews['cleaned_text'])

# Use the Elbow Method to find the optimal number of clusters
sse = []
k_values = range(2, 10)  # Test k from 2 to 10
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)

# Plot the Elbow Method graph
plt.figure(figsize=(8, 5))
plt.plot(k_values, sse, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
# Based on the elbow method, choose k (e.g., 5)
optimal_k = 5
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(X)
reviews['cluster'] = clusters

# Analyze clusters
# Top words for each cluster
terms = tfidf.get_feature_names_out()
for i in range(optimal_k):
    cluster_center = kmeans.cluster_centers_[i]
    top_indices = cluster_center.argsort()[-10:][::-1]  # Get top 10 words
    top_words = [terms[idx] for idx in top_indices]
    print(f"Cluster {i} Top Words: {', '.join(top_words)}")

# Example reviews for each cluster
for i in range(optimal_k):
    print(f"\nCluster {i} Example Reviews:")
    sample_reviews = reviews[reviews['cluster'] == i].sample(3)['reviews.text']
    for review in sample_reviews:
        print(f"- {review}")

# Analyze average rating per cluster
average_ratings = reviews.groupby('cluster')['reviews.rating'].mean()
print("\nAverage Ratings per Cluster:")
print(average_ratings)
```
